## First portion of the data - inversion via Superast

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
plt.figure(figsize=[12, 12])

plt.subplot(2, 2, 1)
stokes = fits.open('qs_ph_disk_center_synth.fits')[0].data
plt.imshow(stokes[:384, :384, 0, 0], origin='lower', cmap='plasma')
plt.title(r'Original data')
plt.colorbar()
plt.subplot(2, 2, 2)
b_c = fits.open('binned_and_convolved_stokes.fits')[0].data
plt.imshow(b_c[:128, :128, 0, 0], origin='lower', cmap='plasma')
plt.title(r'Convolution and binning applied [CB data]')
plt.colorbar()
plt.subplot(2, 2, 3)
invert_ptp = fits.open('modelout_pixel-to-pixel.fits')[0].data
plt.imshow(invert_ptp[:, :, 0], origin='lower', cmap='plasma')
plt.title(r'Pixel-to-pixel inversion on CB data')
plt.colorbar()
plt.subplot(2, 2, 4)
invert_sc = fits.open('modelout_spatially_coupled.fits')[0].data
plt.imshow(invert_sc[:, :, 0], origin ='lower', cmap='plasma')
plt.title(r'Spatially-coupled inversion on CB data')
plt.colorbar()
plt.show()
```

```python
# Flatten the data arrays
stokes_flat = stokes[:384, :384, 0, 0].ravel()
binned_convolved_flat = b_c[:128, :128, 0, 0].ravel()
invert_ptp_flat = invert_ptp[:,:0].ravel()
invert_sc_flat = invert_sc[:,:,0].ravel()

# Create a figure and axis for the histogram
plt.figure(figsize=[8, 6])

# Plot histograms for all data sets
plt.hist(stokes_flat, bins=100, alpha=0.5, label='Original data')
plt.hist(binned_convolved_flat, bins=100, alpha=0.5, label='Convolution and binning applied')
# plt.hist(invert_ptp_flat, bins=100, alpha=0.5, label='Pixel-to-pixel inversion on CB data')
# plt.hist(invert_sc_flat, bins=100, alpha=0.5, label='Spatially-coupled inversion on CB data')

# Add labels and legend
plt.xlabel('')
plt.ylabel(r'N')
plt.title('Histogram of Data from FITS Files')
plt.legend()

# Show the plot
plt.show()
```
